```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/jiadi/OneDrive/Data Analytics/LB_SQL_Project_Data_Job_Analysis/csv_files/Singapore_top_paying_DA_jobs.csv')
```

```python
df
```

```python
df_viz = df[['job_title','salary_year_avg']].copy()
df_viz
```

```python
df_viz = df_viz.groupby('job_title').mean().sort_values('salary_year_avg',ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
plt.figure(figsize=(10,5))
sns.barplot(df_viz, x='salary_year_avg', y='job_title', hue='salary_year_avg', legend=False)
sns.despine()
plt.ylabel('')
plt.xlabel('Yearly Salary (USD)')
plt.title('Data Analyst Jobs with Highest Salaries in Singapore', fontsize= 16, loc='left')
plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'${x/1000:.0f}K'))
plt.show()
```
